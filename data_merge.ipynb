# Soccer players values and their statistics
transfermarkt.de and fbref.com merged data for 2017/18-2019/20 seasons

Data downloaded from Kaggle: https://www.kaggle.com/kriegsmaschine/soccer-players-values-and-their-statistics?select=transfermarkt_fbref_201718.csv

```python
# Player metadata (assume as constant over the years)
id='player'
columns_current = ['nationality','position','squad','age','value','height','foot','league']

# handselected yearly data
years = [17,18,19]
columns_yearly = [
    'games','minutes','shots_total','goal_kicks','goals','assists',
    'cards_yellow','cards_red',
    'passes',
    'corner_kicks','throw_ins',
    'tackles','blocks','interceptions','clearances','touches','dribbles',
    'offsides', 'errors',
    'fouls','fouled',
]

# all columns
columns = columns_current+columns_yearly
```

```python
# helper function to add suffix to yearly columns
def colName(col, year):
    col = str(col)
    if(col == 'player'):
        return col
    else:
        return col + '_' + year
```

```python
data17 = pd.read_csv('transfermarkt_fbref_201718.csv', ';', index_col=id)[columns_yearly]
data18 = pd.read_csv('transfermarkt_fbref_201819.csv', ';', index_col=id)[columns_yearly]
data19 = pd.read_csv('transfermarkt_fbref_201920.csv', ';', index_col=id)
data_current = data19[columns_current]
data19 = data19[columns_yearly]

# rename yearly columns to be able to merge them
data17.columns = [colName(col, '17') for col in data17.columns]
data18.columns = [colName(col, '18') for col in data18.columns]
data19.columns = [colName(col, '19') for col in data19.columns]
```

```python
#check duplicates
data19[data19.index.duplicated(keep=False)]
```

```python
# Sum up duplicates
data17 = data17.groupby([id]).sum()
data18 = data18.groupby([id]).sum()
data19 = data19.groupby([id]).sum()

# remove duplicates in meta data, assume last entry is most recent one
# data_current = data_current.drop_duplicates(subset=['index'], keep='last') # drop d_diplicates doesn't work on index
data_current = data_current.groupby(data_current.index).last()
```

```python
# join data by playername
data = data_current \
    .join(data17) \
    .join(data18) \
    .join(data19)
data.head()
```

```python
data.to_csv('soccer.csv')
```
